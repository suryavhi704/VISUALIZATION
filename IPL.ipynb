```python
import pandas as pd #array operation
import seaborn as sns #for visualisation and plots
import matplotlib.pyplot as plt#for visualisation and plot
```

```python
#loading the ipl dataset.
ipl=pd.read_csv('/content/2649060-matches.csv')
ipl
```

```python
ipl.head()#helps to show the first five records
```

```python
#to show hoe many hows and columns exists
ipl.shape
```

```python
#getting the frequent players with most man of the match awaed
ipl['player_of_match'].value_counts()
```

```python
#getting the top 10 players with most man of the match awaed
ipl['player_of_match'].value_counts()[0:10]
```

```python
#creating a list of top 5 players getting the man of the match award
list(ipl['player_of_match'].value_counts()[0:5].keys())
list(ipl['player_of_match'].value_counts()[0:5])
```

```python
#making a bar plot for the top 5 players with the most man of the match award
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
#getting the frequency of result
ipl['result'].value_counts()
```

```python
#finding the most toss winning team
ipl['toss_winner'].value_counts()
```

```python
#extracting the records where a team won batting first;
betting_first=ipl[ipl['win_by_runs']!=0]
betting_first.keys()
```

```python
#creating a histogram on the column.
plt.figure(figsize=(10,8))
plt.hist(betting_first['win_by_runs'],bins=4,color='g')
plt.xlabel('count of runs...')
plt.ylabel('no. of matches..')
plt.title('NUMBER OF TEAMS WIN WHILE GETTING BAT FIRST ')
plt.show()
```

```python
#MAKING A BAR PLOT FOR TOP THREE TEAMS WITH MOST WINS AFTER BATTING FIRST
betting_first=ipl[ipl['win_by_runs']!=0]
betting_first['winner'].value_counts()[0:3]
plt.figure(figsize=(10,8))
x=list(betting_first['winner'].value_counts()[0:3].keys())
y=list(betting_first['winner'].value_counts()[0:3])
plt.bar(x,y,color=['blue','yellow','red'])
plt.xlabel('Teams')
plt.ylabel('No. of matches..')
plt.title('NUMBER OF TEAMS WIN WHILE GETTING BAT FIRST ')
plt.show()
```

```python
#Making a pie chart
betting_first=ipl[ipl['win_by_runs']!=0]
betting_first['winner'].value_counts()[0:3]
plt.figure(figsize=(10,8))
x=list(betting_first['winner'].value_counts()[0:10].keys())
y=list(betting_first['winner'].value_counts()[0:10])
plt.pie(y,labels=x,autopct='%.0f%%')
plt.show()
```
